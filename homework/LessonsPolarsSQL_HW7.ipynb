# HW7. Группировка и агрегация данных

Импорт необходимых библиотек

```python
import polars as pl
import random
```

Генерация данных

```python
# Устанавливаем фиксированный SEED для воспроизводимости
SEED = 100
random.seed(SEED)

# Объём данных
n = 55000

# id деревьев
tree_ids = list(range(1, n + 1))
# Виды деревьев
species = [random.choice(["ель", "берёза", "дуб", "клён", "сосна", "липа", None]) for _ in range(n)]
# Местоположение
location = [random.choice(["парк", "лес", "улица", "сад", None, "территория школы"]) for _ in range(n)]
# Возраст
age_years = [random.choice([10, 15, 20, 25, 30, 40, 50, 80, 120, 150, None]) for _ in range(n)]
# Высота в метрах
height_m = [random.choice([5.2, 8.7, 12.3, 15.0, 18.5, 22.0, 25.5, 30.0, None]) for _ in range(n)]
# Диаметр ствола в см
trunk_diameter_cm = [random.choice([15, 20, 25, 30, 35, 40, 50, 60, 80, 100, None]) for _ in range(n)]
# Диаметр кроны в метрах
crown_diameter_m = [random.choice([3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0, None]) for _ in range(n)]
# Индекс здоровья (0–100)
health_index = [random.choice([95, 88, 80, 75, 70, 60, 50, 40, 30, None]) for _ in range(n)]

# Создание DataFrame
df_trees = pl.DataFrame({
    "tree_id": tree_ids,
    "species": species,
    "age_years": age_years,
    "height_m": height_m,
    "trunk_diameter_cm": trunk_diameter_cm,
    "crown_diameter_m": crown_diameter_m,
    "health_index": health_index,
    "location": location
})
```

- `species` - str, Вид дерева;

- `height_m` - f64, Высота дерева в метрах;
- `trunk_diameter_cm` - i64, Диаметр ствола в сантиметрах;
- `crown_diameter_m` - f64, Диаметр кроны в метрах;
- `health_index` - i64, Индекс здоровья;
- `location` - str, Тип местоположения;
