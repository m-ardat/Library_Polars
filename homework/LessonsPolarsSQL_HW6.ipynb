Импорт необходимых библиотек

```python
import polars as pl
from datetime import datetime, timedelta
import random
```

Генерация данных

```python
# Устанавливаем фиксированный seed для воспроизводимости
SEED = 101
random.seed(SEED)

# Генерация данных
n = 100001
customer_ids = list(range(1, n + 1))

# Все значения для age_raw строки
ages = [str(random.choice([25, 30, 35, 40, 45, 50, 72, 43, 67, 18, None, "unknown", "25-30"])) for _ in range(n)]
# status_codes
status_codes = [random.choice(["active", "inactive", "pending", "suspended", None]) for _ in range(n)]
# scores - строки
scores = [str(random.choice([85.5, 90.0, 78.2, 100.0, None, "N/A"])) for _ in range(n)]
# purchase_amount — числа и None
purchase_amounts = [random.choice([100.0, 250.0, 0.0, None]) for _ in range(n)]
# names - строки
names = [str(random.choice(["  Alice ", "Bob", "Charlie", "  ", "David", "Eve", "  Frank  ", "Grace", "null", "Helen"])) for _ in range(n)]

df_customers = pl.DataFrame({
    "customer_id": customer_ids,
    "name": names,
    "age_raw": ages,
    "status": status_codes,
    "score_raw": scores,
    "purchase_amount": purchase_amounts
})
```
