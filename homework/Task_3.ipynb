
**Задача №3. Продажи магазина.**

Есть две таблицы: таблица продаж `df_sales`, в которой содержится информация о дате покупки, номере магазина, артикуле, который продали и количество проданных штук данного артикула. Вторая таблица `df_prices` - справочник цен, в котором есть информация по ценам всех артикулов из таблицы `df_sales`.

Посчитайте сумму продаж по 100му магазину за 1 января 2013 года. Ответ округлите до сотых.


Импортируем необходимые библиотеки

```python
import polars as pl
import random
from datetime import datetime, timedelta
```

Источник данных

```python
# Для получения одних и теж результатов
random.seed(10)

# Параметры
n_sales = 5000000
n_articles = 1000
shops = [x for x in range(1, 200)]

# Генерация Prices: каждый артикул — уникальный, одна цена
articles = [f"A{i+1}" for i in range(n_articles)]
prices_data = [{"art": art,"price": round(random.uniform(5, 200), 2)} for art in articles]

df_prices = pl.DataFrame(prices_data)

# Генерация Sales
start_date = datetime(2012, 12, 25)
end_date = datetime(2013, 1, 5)
date_range = [start_date + timedelta(days=i) for i in range((end_date - start_date).days + 1)]

sales_data = []
for _ in range(n_sales):
    dt = random.choice(date_range) + timedelta(
        hours=random.randint(0, 23),
        minutes=random.randint(0, 59),
        seconds=random.randint(0, 59)
    )
    shop = random.choice(shops)
    art = random.choice(articles)
    qty = random.randint(-5, 20)
    sales_data.append({
        "datetime": dt,
        "shop": shop,
        "art": art,
        "quantity": qty
    })

df_sales = pl.DataFrame(sales_data).with_columns(pl.col("datetime").dt.replace_time_zone(None))
```
