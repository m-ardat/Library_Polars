Импорт необходимых библиотек

```python
import polars as pl
from datetime import datetime, timedelta
import random
import numpy as np
```

Генерация данных

```python
# Устанавливаем фиксированный seed для воспроизводимости
SEED = 100
random.seed(SEED)
np.random.seed(SEED)
# Генерация данных
n = 12345  # количество строк
order_ids = [f"ORD-{(datetime(2025, 11, 1) + timedelta(days=i % 30)).strftime('%Y%m%d')}-{str(i).zfill(5)}" for i in range(n)]
names = ["  Владимир Иванов ", "Том Пират ", "Анастасия Кузнецова", " Владимир ", "Михаил Хлебов", " Валера Тихий", "Мария Карпова"]
products = ["Ноутбук", "Мышь", "Клавиатура", "Монитор", "Телефон", "Камера"]
statuses = ["pending", "shipped", "delivered", "cancelled"]

df_orders = pl.DataFrame({
    "order_id": order_ids,
    "customer_name": [random.choice(names) for _ in range(n)],
    "product_name": [random.choice(products) for _ in range(n)],
    "quantity": [random.randint(1, 10) for _ in range(n)],
    "price": [round(random.uniform(10.0, 1000.0), 2) for _ in range(n)],
    "order_date": [datetime(2025, 11, 1) + timedelta(days=random.randint(0, 30), hours=random.randint(0, 23)) for _ in range(n)],
    "status": [random.choice(statuses) for _ in range(n)]
})
```
